```python
# üìå Step 1 ‚Äì Install the Together AI SDK
!pip install together

# üìå Step 2 ‚Äì Import and Initialize
from together import Together

# Your Together API Key (KEEP THIS PRIVATE)
TOGETHER_API_KEY = "11f4c4cb73624eaa87152e5e21c85f3b3b2dfc68e9cde7160ad3093fec57f9ea"
client = Together(api_key=TOGETHER_API_KEY)

print("ü§ñ DeepCogito v2 Chat (type 'exit' to quit)\n")

# üìå Step 3 ‚Äì Chat loop
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("üëã Ending chat. Have a great day!")
        break

    response = client.chat.completions.create(
        model="deepcogito/cogito-v2-preview-llama-70B",
        messages=[{"role": "user", "content": user_input}],
        enable_thinking=True
    )

    print("\nDeepCogito v2:", response.choices[0].message.content, "\n")
```
